```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt
import plotly.express as ps;

data = pd.read_csv('/content/apple_products (1).csv',encoding='utf-8-sig')
print(data.columns)

```

```python
#DATA CLEANING
print(data.info())
print('/n',data.isnull().sum())
```

```python
#DESCRIPTIVE ANALYSIS
print(data['Sale Price'].describe())
```

```python
#iphone sale aanalysis in INDIA
highest_rated = data.sort_values(['Star Rating'],ascending=False)
highest_rating = highest_rated.head(10);
print(highest_rating['Product Name'])
x = highest_rating['Product Name']
y = highest_rating ['Number Of Reviews']
plt.xticks(rotation=90)

plt.bar(x,y)
plt.xlabel('Product Name')
plt.ylabel('Number Of Reviews')

plt.legend()
plt.show()
```

```python
#Sales price
x  =  highest_rating['Product Name']
y =  highest_rating['Sale Price']
plt.xlabel('Sales Price')
plt.ylabel('Product Name')
plt.scatter(x,y)
plt.xticks(rotation=90)

plt.show()
```
